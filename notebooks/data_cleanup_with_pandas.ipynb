```python
#### filtering data with str.contains ######
```

```python
# The str method has something called contains which returns true or false.
# These true or false can be passed to loc and second arugment to get all or the limited columns
data.loc[ data.medium.str.contains('Graphite'), :]
```

```python
# contains is case sensitive. can be overriden with case=False
data.loc[data.medium.str.contains('Graphite', case=False)]
```

```python
# contains can take regular expression args as well
data.loc[data.medium.str.contains('(?i)Graphite', regex=True), :]
```

```python
# but this cannot run on numeric values, so they should be converted to string first
data.year.astype(str).str.contains('12')
```

```python
# this could be passed to loc and then get the full dataset
data.loc[ data.year.astype(str).str.contains('17'), :]
```

```python
# na = False has filtered the NaN values
data.loc[ data.dimensions.str.contains('support', na=False), :]
```
