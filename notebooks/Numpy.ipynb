```python
#### Indexing and slicing ####

# indexing is same as how it's in the python list
ari = np.arange(11)**2
ari
```

```python
ar2d = np.arange(15).reshape(5,3)
print(ar2d, "\n", ar2d[0],"\n") #for a 2d array the first row is accesible with row,column method. So ar2d[0] will print the whole row
print(ar2d[0,2],"\n") # accessing specific element with row.column
```

```python
#specifiying range for rows and columns
ar2d[0:2, 1:2] # is 0:2 rows and 2:4 range for columns 
```

```python
ar2d[:,1:3] # all the rows but 1:3 range for columns
```

```python
# negative index for multi dimensional arrays
ar2d[-1:,:-1]
```

```python
## iterating over 2d+ arrays
# This iterates through every number in the matrix
for i in ar2d:
    print(i)
```

```python
# using flatten function you get every element in the array one by one
# it takes argument order, possible values = F - columnar printing, column1, 2, 3 etc.
# returns the results of the flattened list
for i in ar2d.flatten(order='F'):
    print(i)
```

```python
#iterating with nditer - this iterated in row major form again. order=F can be used to column major
# does not retuen anything, just allwos you to interate through
for i in np.nditer(ar2d):
    print(i)
```

```python
print(ar2d)
# using nditer to get inner most loop ???? This has created transpose of the matrix
for i in np.nditer(ar2d, order='F', flags = ['external_loop']):
    print(i)
    
print(ar2d)
```

```python
#nd iter in readwrite mode
for i in np.nditer(ar2d):
    print(i)

ar2d

# updating 2d array in place
for i in np.nditer(ar2d, op_flags=['readwrite']):
    i[...] = i * i 

ar2d
```
